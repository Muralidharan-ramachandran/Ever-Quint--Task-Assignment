```python
def all_optimal_plans(n):
    available_time = n - 1
    plans = []

    # Step 1: Generate all valid plans + remaining time
    for C in range(0, available_time // 10 + 1):
        remaining1 = available_time - 10 * C

        for T in range(0, remaining1 // 5 + 1):
            remaining2 = remaining1 - 5 * T

            for P in range(0, remaining2 // 4 + 1):

                time_used = 10*C + 5*T + 4*P
                RT = n - time_used   # remaining operational time

                # Step 2: Calculate profit
                profit = (
                    RT * T * 1500 +
                    RT * P * 1000 +
                    RT * C * 2000
                )

                plans.append((T, P, C, RT, profit))

    # Step 3: Find maximum profit
    max_profit = max(plan[4] for plan in plans)

    # Step 4: Return all plans that match max profit
    optimal = [plan for plan in plans if plan[4] == max_profit]

    return optimal

n = int(input("Enter the value of n: "))
optimal = all_optimal_plans(n)
print(f"All optimal ways for n={n}:\n")
for plan in optimal:
    T, P, C, RT, profit = plan
    print(f"T={T}, P={P}, C={C}, RT={RT}, profit={profit}")
```
